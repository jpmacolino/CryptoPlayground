```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
btc = pd.read_csv('cb_BTCUSD_daily.csv', index_col='date', parse_dates=[0])
```

```python
btc.head()
```

```python
btc.drop('unix', axis=1, inplace=True)
```

```python
btc.info()
```

```python
btc.index = pd.to_datetime(btc.index)
```

```python
btc = btc.sort_values('date')
```

Plotting Some Basic Figures

```python
plt.style.use('seaborn')
fig, ax = plt.subplots()
plt.plot(btc.index[269:], btc.high[269:])
plt.plot(btc.index[269:], btc.low[269:])
plt.title('High and Low BTC Prices (Last 30 Days)')
plt.ylabel('Price')
plt.xlabel('Date')
plt.legend(['High', 'Low'])
plt.xticks(rotation=45);
```
